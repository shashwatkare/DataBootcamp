````python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_dir = "~/UCBERKDATA/GitLab/UCBER201805DATA2-Class-Repo/00-Homework/04-Numpy-Pandas/Instructions/HeroesOfPymoli/"
file_to_load = file_dir+"Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
````

## Player Count

* Display the total number of players


````python
total_players = len(purchase_data['SN'].unique())
pd.DataFrame([{'Total Players': total_players}])
````

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


````python
num_of_unique_items = len(purchase_data['Item ID'].unique())
avg_price = purchase_data.groupby('Item ID')['Price'].mean().mean()
#avg_price = purchase_data['Price'].mean()
num_of_purchases = len(purchase_data['Purchase ID'].unique())
total_revenue = purchase_data['Price'].sum()

PADF = pd.DataFrame({'Number of Unique Items': [num_of_unique_items], 'Average Price': [avg_price],\
               'Number of Purchases': [num_of_purchases], 'Total Revenue': [total_revenue]})
PADF['Average Price'] = PADF['Average Price'].map("${:.2f}".format)
PADF['Total Revenue'] = PADF['Total Revenue'].map("${:.2f}".format)
PADF
````

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


````python
upg = purchase_data.groupby('SN')['Gender'].value_counts().index.tolist()
Player_Genders = pd.DataFrame({'unique_player_gender': [x[1] for x in upg]})
gender_list = Player_Genders['unique_player_gender'].value_counts().index.tolist()
player_gender_count = Player_Genders['unique_player_gender'].value_counts().tolist()
player_gender_percentage = [100.*x/len(Player_Genders['unique_player_gender']) for x in player_gender_count]
PAGDDF = pd.DataFrame({' ':gender_list, 'Percentage of Players':player_gender_percentage,\
                        'Total Count': player_gender_count})
PAGDDF = PAGDDF.set_index(' ')
PAGDDF['Percentage of Players'] = PAGDDF['Percentage of Players'].map("{:.2f}%".format)
PAGDDF
````

````python
M2_genderlist = purchase_data['Gender'].value_counts().index.tolist()
M2_gendercount = purchase_data['Gender'].value_counts().values.tolist()
M2_percentage = [100.*x/purchase_data['Gender'].value_counts().sum() for x in M2_gendercount]

PAGDDF2 = pd.DataFrame({' ':M2_genderlist, 'Percentage of Players':M2_percentage,\
                       'Total Count': M2_gendercount})
PAGDDF2 = PAGDDF2.set_index(' ')
PAGDDF2['Percentage of Players'] = PAGDDF2['Percentage of Players'].map("{:.2f}%".format)
PAGDDF2
````


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

````python
PAGDDF3 = pd.DataFrame({'Gender':purchase_data['Gender'].value_counts().index.tolist(),\
                        'Purchase Count':purchase_data['Gender'].value_counts().values.tolist()})
PAGDDF3 = PAGDDF3.set_index('Gender')
PAGDDF3['Average Purchase Price'] = purchase_data.groupby('Gender')['Price'].mean()
PAGDDF3['Total Purchase Value'] = purchase_data.groupby('Gender')['Price'].sum()
PAGDDF3['Normalized Totals'] = PAGDDF3['Total Purchase Value']/PAGDDF3['Purchase Count']
PAGDDF3['Average Purchase Price'] = PAGDDF3['Average Purchase Price'].map("${:.2f}".format)
PAGDDF3['Total Purchase Value'] = PAGDDF3['Total Purchase Value'].map("${:,.2f}".format)
PAGDDF3['Normalized Totals'] = PAGDDF3['Normalized Totals'].map("${:,.2f}".format)
PAGDDF3
````

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


````python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
bin_result = purchase_data.groupby(pd.cut(purchase_data['Age'],age_bins,labels=group_names))['Item ID'].count()
AGEDF = pd.DataFrame({'Total Count': bin_result})
AGEDF['Total Count'].sum()
AGEDF['Percentage of Players'] = 100.*AGEDF['Total Count']/AGEDF['Total Count'].sum()
AGEDF['Percentage of Players'] = AGEDF['Percentage of Players'].map("{:,.2f}%".format)
AGEDF
````

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

````python
AGEDF2 = AGEDF.drop(columns=['Percentage of Players'])
AGEDF2.rename(columns={'Total Count':'Purchase Count'}, inplace=True)
AGEDF2['Average Purchase Price'] = purchase_data.groupby(pd.cut(purchase_data['Age'],age_bins,labels=group_names))['Price'].mean()
AGEDF2['Total Purchase Value'] = purchase_data.groupby(pd.cut(purchase_data['Age'],age_bins,labels=group_names))['Price'].sum()
AGEDF2['Normalized Totals'] = AGEDF2['Total Purchase Value']/AGEDF2['Purchase Count']
AGEDF2['Average Purchase Price'] = AGEDF2['Average Purchase Price'].map("${:,.2f}".format)
AGEDF2['Total Purchase Value'] = AGEDF2['Total Purchase Value'].map("${:,.2f}".format)
AGEDF2['Normalized Totals'] = AGEDF2['Normalized Totals'].map("${:,.2f}".format)
AGEDF2
````

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



````python
TSDF = pd.DataFrame({'Purchase Count':purchase_data.groupby('SN')['SN'].count()})
TSDF['Average Purchase Value'] = purchase_data.groupby('SN')['Price'].mean()
TSDF['Total Purchase Value'] = purchase_data.groupby('SN')['Price'].sum()
TSDF = TSDF.sort_values(by=['Total Purchase Value'],ascending=False)
TSDF['Average Purchase Value'] = TSDF['Average Purchase Value'].map("${:,.2f}".format)
TSDF['Total Purchase Value'] = TSDF['Total Purchase Value'].map("${:,.2f}".format)
TSDF.head(5)
````

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



````python
PIDF = pd.DataFrame({'Purchase Count':purchase_data.groupby(['Item ID','Item Name'])['Item ID'].count()})
PIDF['Item Price'] = purchase_data.groupby(['Item ID','Item Name'])['Price'].mean()
PIDF['Total Purchase Value'] = purchase_data.groupby(['Item ID','Item Name'])['Price'].sum()
PIDF2 = PIDF.copy()
PIDF['Total Purchase Value'] = PIDF['Total Purchase Value'].map("${:,.2f}".format)
PIDF['Item Price'] = PIDF['Item Price'].map("${:,.2f}".format)
PIDF = PIDF.sort_values(by=['Purchase Count'],ascending=False)
PIDF.head(5)
````

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



````python
PIDF2 = PIDF2.sort_values(by=['Total Purchase Value'],ascending=False)
PIDF2['Total Purchase Value'] = PIDF2['Total Purchase Value'].map("${:,.2f}".format)
PIDF2['Item Price'] = PIDF2['Item Price'].map("${:,.2f}".format)
PIDF2.head(5)
````
